```python
# CHECKING THE NUMBER OF CLUSTER 
```

```python
from sklearn.cluster import KMeans
```

```python
# WCSS = within cluster sum of square 
```

```python
wcss = [1000,2000,3000,4000,5000,6000,7000,8000,9000]
for i in range(1,11):
    kmean = KMeans(n_clusters=i, init='k-means++',random_state=42)
    kmean.fit(X)
wcss.append(kmean.inertia_)
```

# MAKING THE CLUSTER 

```python
from sklearn.cluster import KMeans

# Now create the KMeans object
km = KMeans(n_clusters=5,init='k-means++', random_state=0)
y = km.fit_predict(X)
```

```python
y
```

```python
X = pd.DataFrame(X)
```

```python
data["predict"] = km.predict(X)
sns.pairplot(data=data, hue='predict')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(8, 8))

y_list = y.tolist() if hasattr(y, 'tolist') else list(y)
unique_labels = set(y_list)
colors = ['green', 'red', 'yellow', 'violet', 'blue', 'orange', 'purple', 'cyan', 'magenta', 'brown']
for i, label in enumerate(unique_labels):
    #
    indices = [j for j, val in enumerate(y_list) if val == label]
    
    
    X_array = np.array(X)
    cluster_points = X_array[indices]
    
    
    if len(cluster_points) > 0:
        x_coords = cluster_points[:, 0]  # First column of coordinates
        y_coords = cluster_points[:, 1]  # Second column of coordinates
        ax.scatter(x_coords, y_coords, s=50, 
                  c=colors[i % len(colors)], label=f'cluster_{label}')


ax.legend()


plt.show()
```
